```python
import torch
from transformers import AutoTokenizer,AutoModelForCausalLM
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
MODEL = "stanford-crfm/BioMedLM"
```

```python
device = torch.device("cuda")

tokenizer = GPT2Tokenizer.from_pretrained("stanford-crfm/BioMedLM")

model = GPT2LMHeadModel.from_pretrained("stanford-crfm/BioMedLM").to(device)
```

```python
def model_inference(input_text):
    device = torch.device("cuda")

    input_ids = tokenizer.encode(
        input_text, return_tensors="pt"
    ).to(device)

    sample_output = model.generate(input_ids, do_sample=True, max_length=100, top_k=50)

    print("Output:\n" + 100 * "-")
    print(tokenizer.decode(sample_output[0], skip_special_tokens=True))
    return True
```

```python
model_inference("medicine for gut health")
```

```python
model_inference(input_text="hair fall solutions")
```
